## 1) Import Essentials

```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

## 2) Initialize init, position, and velocity of particles

```python
class Dynamics:
    def __init__(self):
        self.dim = 2 #dimensions
        self.num = 60 #number of particles
        self.box_len = 12
        self.mass = 1 #mass of particles
        self.temp = 300
        self.k = 1
        self.positions = np.zeros((self.num, self.dim)) #particles begin at position zero
        self.velocities = np.zeros((self.num, self.dim))
    #fx for position
    def position_particles(self):
        self.positions = np.random.random_sample((self.num, self.dim))*self.box_len #random positons for the different particles, range from 0-12
    #fx for velocity
    def velocity_particles(self):
        #minimization - gaps between particles
        std_dev = (self.mass/(self.k*self.temp))**.5
        self.velocities = np.random.normal(loc=0,scale =std_dev,size=(self.num,self.dim))
    #fx for plotting molecules
    def plot_particles(self):
        plt.figure(figsize=(5,5))
        axis = plt.gca()

        #drawing area boundaries
        axis.set_xlim(-10,self.box_len+10)
        axis.set_ylim(-10,self.box_len+10)
        
        #add circles to particles' locations
        for i in range(self.num):
            axis.add_patch(plt.Circle(self.positions[i], radius = .5,
                                     linewidth=1.5, edgecolor = "black"))
```

## 3) Print out array of molecules' positions and velocities

```python
#define functions to variable
mol_dynamics = Dynamics()

mol_dynamics.position_particles()
mol_dynamics.velocity_particles()
```
